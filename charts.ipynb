```python
# ScatterPlot of Binding Affinity vs Tanimoto Similarity to DHT with Regression Line and R^2 and P-Value
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Load the data and set the index
df = pd.read_csv('results.csv')  
df = df.set_index('ID')

# Calculate the regression for Binding Affinity vs Tanimoto Similarity
formula = 'Q("Binding Affinity") ~ Q("Tanimoto Similarity")'
model = smf.ols(formula=formula, data=df)
results = model.fit()

# Print the summary of the regression results
print(results.summary())
print("\n" + "="*80 + "\n") 

# Get the values to display on the plot
r_squared = results.rsquared
p_value_x = results.pvalues['Q("Tanimoto Similarity")'] 
intercept = results.params['Intercept']
slope = results.params['Q("Tanimoto Similarity")']

# Create the scatter plot with regression line
fig, ax = plt.subplots()
ax.scatter(df['Tanimoto Similarity'], df['Binding Affinity'])
# Plots the regression line
ax.plot(df['Tanimoto Similarity'], slope * df['Tanimoto Similarity'] + intercept, color='red', label='Regression Line')

# Add the equation and statistics to the plot (only first 3 decimal places)
equation = f'Equation: y = {slope:.3f}x + {intercept:.3f}'
r2_text = f'$R-Squared = {r_squared:.3f}$'
p_value_text = f'$p$-value = {p_value_x:.3f}'

full_text = f'{equation}\n{r2_text}\n{p_value_text}'

# Define the limits for the axes
plt.xlim(-0.05, 1.05)  
plt.ylim(-13, 6)  

# Add the text box with the equation and statistics
ax.text(0.95, 0.95, full_text, transform=ax.transAxes, fontsize=12,
        verticalalignment='top', horizontalalignment='right', bbox=dict(boxstyle='round,pad=0.5', fc='wheat'))

# Set labels and title
plt.xlabel('Tanimoto Similarity to DHT')
plt.ylabel('Binding Affinity')
plt.title('Binding Affinity vs Tanimoto Similarity to DHT')
plt.show()
```

```python
# ScatterPlot for DHT and AV6, showing outliers in the graph and labeling them
import pandas as pd
import matplotlib.pyplot as plt

# Load the data and index by ID
df = pd.read_csv('results.csv')  
df = df.set_index('ID')

# Specify which IDs to show in the scatter plot
ids_to_show = ['DHT', 'AV6']
df_filtered = df.loc[ids_to_show]

# Create the plot
fig, ax = plt.subplots()
ax.scatter(df_filtered['Tanimoto Similarity'], df_filtered['Binding Affinity'])

# Add the labeling for the IDs
for id in ids_to_show:
        x = df_filtered.loc[id, 'Tanimoto Similarity']
        y = df_filtered.loc[id, 'Binding Affinity']
        ax.annotate(id, (x, y), textcoords="offset points", xytext=(0,10), ha='center')

# Specify the limits for the axes
plt.xlim(-0.05, 1.05)  
plt.ylim(-13, 6)  

# Add labels and title
plt.xlabel('Tanimoto Similarity to DHT')
plt.ylabel('Binding Affinity')
plt.title('Binding Affinity vs Tanimoto Similarity to DHT')
plt.show()
```
